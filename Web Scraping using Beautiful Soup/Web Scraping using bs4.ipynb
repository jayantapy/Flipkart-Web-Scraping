```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
print(soup.prettify())
```

### Get the products

```python
product = soup.find_all('div',class_= '_4rR01T')
product
```

```python
products = []
for i in range(0,len(product)):
    products.append(product[i].get_text())
```

```python
products
```

```python
len(products)
```

### Get the ratings

```python
rating = soup.find_all('div',class_ = "_3LWZlK")
rating
```

```python
ratings = []
for i  in range(len(rating)):
    ratings.append(rating[i].get_text())
```

```python
ratings
```

```python
ratings = ratings[:24]
len(ratings)
```

### Get the price

```python
price = soup.find_all('div',class_="_30jeq3 _1_WHN1")
price
```

```python
prices = []
for i in range(len(price)):
    prices.append(price[i].get_text())
```

```python
prices
```

```python
len(prices)
```

```python
description = soup.find_all('ul',class_="_1xgFaf")
description
```

```python
descriptions=[]
for i in range(0, len(description)):
    descriptions.append(description[i].get_text())

descriptions
```

```python
len(descriptions)
```

```python
lst = []
for i in descriptions:
    lt = i.split()
    n = lt.index("DDR4")
    ram = ' '.join(lt[n-2:n+1])
    if len(ram.split()[0]) == 5:
        ram = " ".join(list(ram)[ram.find(")")+1:])
    else:
        ram = ram.split()
        ram = [ram[0][-1]] + ram[1:]
        ram = ' '.join(ram)
    lst.append(ram)
```

```python
lst
```

```python
RAM = []
for i in lst:
    RAM.append(i[0])
```

```python
RAM
```

### Converting scraped data into dataframe

```python
import pandas as pd
```

```python
df=pd.DataFrame()
df['Product']=products
df['Description']=descriptions
df['Rating']=ratings
df['Price']=prices
df['RAM'] = RAM
```

```python
df
```

## Laptops with RAM of 8 GB
